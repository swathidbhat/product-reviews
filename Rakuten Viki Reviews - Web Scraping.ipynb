```python
# CONTEXT
# I wanted to identify pain points experienced by Rakuten Viki's users 
# I wanted to analyse user reviews left on Google Play Store to understand pain points experienced by users of Viki's android app.
```

```python
import pandas as pd
```

```python
# Found this scraper to get reviews https://pypi.org/project/google-play-scraper/ . It accepts app merchant ID as input. 
# pip install google-play-scraper
```

```python
from google_play_scraper import app, reviews, reviews_all, Sort
```

```python
viki_user_reviews = reviews_all('com.viki.android')
```

```python
viki_user_reviews
```

```python
user_review_df = pd.DataFrame.from_dict(viki_user_reviews)
user_review_df.head()
```

```python
user_review_df.columns
user_review_df.count()
```

```python
# filtering for reviews that rate Grab Merchant app not more than 3 stars. 
bad_reviews_df = user_review_df[user_review_df['score'] < 4] 
bad_reviews_df.head()
```

```python
bad_reviews_df.count()
```

```python
# filtering for reviews that have been upvoted at least 10 times, to indicate the scale of an issue. 
pop_bad_reviews_df = bad_reviews_df[bad_reviews_df['thumbsUpCount'] > 99]
pop_bad_reviews_df.head()
```

```python
pop_bad_reviews_df.count()
```

```python
pop_bad_reviews_df = pop_bad_reviews_df.sort_values(by=['at', 'thumbsUpCount'], ascending=False)
pop_bad_reviews_df
```

```python
pop_bad_reviews_df.to_csv('popular-viki-user-reviews.csv')
```
